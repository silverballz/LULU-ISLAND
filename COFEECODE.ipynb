##Lets make a scatter plot for Coding Hours vs CoffeeCupsPerDay

```python
df.plot(kind='scatter', x='CodingHours', y='CoffeeCupsPerDay', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

##Lets make a bar histogram for Coffee Type and their frequency

```python
df.groupby('CoffeeType').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

###Histogram for CodingHours vs Frequency

```python

df['CodingHours'].plot(kind='hist', bins=20, title='CodingHours')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

##Lets check if there is a correlation between Coding Hours and CoffeeCupsPerDay

```python
p_correlation = df["CodingHours"].corr(df["CoffeeCupsPerDay"])

# Print correlation coefficient
print("Correlation of 'CodingHours' and 'CoffeeCupsPerDay':", p_correlation)
```

##As the Correlation constant is less than 0.5, then lets check the correlation for a particular country say Lebanon

```python
# Declare variable `df_lebanon` by subsetting `df`
df_lebanon = df[df["Country"] == "Lebanon"]

# Print object type, shape, and head

df_lebanon.head()
```

```python
# Create scatter plot of "CodingHours" vs "CoffeeCupsPerDay" in Lebanon
plt.scatter(x=df_lebanon["CodingHours"], y=df_lebanon["CoffeeCupsPerDay"])

# Add x-axis label
plt.xlabel("CodingHours")

# Add y-axis label
plt.ylabel("CoffeeCupsPerDay")

# Add title
plt.title("Lebanon: CodingHours vs. CoffeeCupsPerDay");
```

```python
# Calculate correlation of "CodingHours" and "CoffeeCupsPerDay" in `df_lebanon`
p_correlation = df_lebanon["CodingHours"].corr(df_lebanon["CoffeeCupsPerDay"])

# Print correlation coefficient
print("Correlation of 'CodingHours' and 'CoffeeCupsPerDay' (Lebanon):", p_correlation)
```

#As the correlation is still small so we can conclude that CodingHours isnt influenced by CofeeCupsPerDay in Lebanon atleast.
